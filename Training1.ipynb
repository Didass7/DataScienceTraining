```python
import mlcroissant as mlc
import pandas as pd

# Fetch the Croissant JSON-LD
croissant_dataset = mlc.Dataset('https://www.kaggle.com/datasets/navjotkaushal/coffee-sales-dataset/croissant/download')

# Check what record sets are in the dataset
record_sets = croissant_dataset.metadata.record_sets
print(record_sets)

# Fetch the records and put them in a DataFrame
df = pd.DataFrame(croissant_dataset.records(record_set=record_sets[0].uuid))
df.head()
```

```python
media_vendas_por_mes = df.groupby('Coffe_sales.csv/Month_name')['Coffe_sales.csv/money'].mean()
print(media_vendas_por_mes)
```

```python
vendas_por_mes = df.groupby('Coffe_sales.csv/Month_name')['Coffe_sales.csv/money'].sum()
print(vendas_por_mes)
```

```python
import matplotlib.pyplot as plt

# Criar o gráfico de linhas
vendas_por_mes.plot(kind='line', figsize=(10, 6))
plt.title('Vendas Totais por Mês')
plt.xlabel('Mês')
plt.ylabel('Vendas Totais')
plt.grid(True)
plt.show()
```

```python
# Converter a coluna de data para o formato datetime (se ainda não o fez)
df['Coffe_sales.csv/Date'] = pd.to_datetime(df['Coffe_sales.csv/Date'])

# Criar uma nova coluna com o número do mês
df['month_num'] = df['Coffe_sales.csv/Date'].dt.month
```

```python
vendas_por_mes_ordenado = df.groupby('month_num')['Coffe_sales.csv/money'].sum()
```

```python
import matplotlib.pyplot as plt

vendas_por_mes_ordenado.plot(kind='line', figsize=(10, 6))
plt.title('Vendas Totais por Mês (Ordenado)')
plt.xlabel('Mês (Número)')
plt.ylabel('Vendas Totais')
plt.grid(True)
plt.show()
```
