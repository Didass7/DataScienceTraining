```python
#Quantidade de Valores Nulos na coluna Idade
df.age.isnull().sum()
```

```python
#Quantidade de Valores Nulos na coluna Deck
df.deck.isnull().sum()
```

```python
#Calcula a mediana das Idades
#É mais reliable para preencher os nulos
df.age.median()
```

```python
#Calcula a média das Idades
#É menos reliable para preencher os nulos
df.age.mean()
```

```python
#Substitui os valores das Idades Nulas pelo valor da Mediana de todas as idades (28)
df['age']=df['age'].fillna(df['age'].median())
```

```python
df
```

```python
df.info()
```

```python
#Como a coluna deck é do tipo category é preciso adicionar a categoria "Unknown" para poder substituir os nulos
df['deck'].value_counts()
```

```python
#Adiciona a categoria à coluna deck
df['deck'] = df['deck'].cat.add_categories('Unknown')
```

```python
#Substitui os Nulos para a categoria "Unknown"
df['deck'] = df['deck'].fillna('Unknown')
```

```python
# Nº de Pessoas por classe
df['pclass'].value_counts()
```

```python
# Ver a média de idade por classe
df.groupby('pclass')['age'].mean()
```

```python
#Ver a tarifa média (fare) por classe e por sexo:
df.groupby(['pclass','sex'])['fare'].mean()
```

```python
# Encontrar passageiros com mais de 50 anos:
df.query('survived==1 & pclass==3')
```

```python
#Selecionar apenas as colunas sex, age, fare para os passageiros do sexo feminino.
df.query('sex == "female"')[['age','fare','sex']]
```

```python
#Contar nº de sobreviventes por sexo:
df.groupby('sex')['survived'].sum()
```
