```python
#grÃ¡fico de linhas mostrando a mÃ©dia de idade por classe, separado entre sobreviventes e nÃ£o sobreviventes

media_idade = df.groupby(['pclass', 'survived'])['age'].mean().reset_index()

nao_sobreviventes = media_idade[media_idade['survived'] == 0]
sobreviventes = media_idade[media_idade['survived'] == 1]

plt.plot(
    nao_sobreviventes['pclass'], 
    nao_sobreviventes['age'],
    marker='o',
    color='red',
    label='NÃ£o Sobreviventes'
)

plt.plot(
    sobreviventes['pclass'], 
    sobreviventes['age'],
    marker='o',
    color='green',
    label='Sobreviventes'
)

plt.title('MÃ©dia de Idade por Classe, Separada por Status de SobrevivÃªncia')
plt.xlabel('Classe')
plt.ylabel('MÃ©dia de Idade')
plt.xticks([1, 2, 3], ['1Âª Classe', '2Âª Classe', '3Âª Classe'])
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
```
